<h3>Totalt population in regions, Most to least</h3>

<h3>Step 4: Fråga och svara på 4 intressanta frågor om datan</h3>

<h5>Q1, Which guest country was getting the most resonses?</h5>

```python
norev = reviews.loc[:,['GuestCountry', 'PropertyResponse']]
norev.drop(norev[norev['PropertyResponse']=='-'].index, inplace = True)
nr = norev.groupby(['GuestCountry']).count().sort_values('PropertyResponse',ascending=False)
nr
```

```python
norev.groupby(['GuestCountry']).count().plot()
plt.xlabel('Countries')
plt.ylabel('Responses')
plt.xticks(rotation=90)
plt.legend(["Amount of responses"])
plt.show()
```

<h4>I created a plot that draw lines between all the values in the slices and grouping of amount of responses to visiting guest by country in the dataset. To make it more understandable I've changed the labels of the y and x - axis. I renamed the legend asweel to get a nicer look to it. <h5>Q2, From which 5 countries were guests giving the highest score of their stay</h5>

```python
highscore = reviews.loc[:,['GuestCountry', 'Score']]
hs = highscore.loc[(highscore['Score'] == 10)].groupby(by=['GuestCountry']).count().sort_values('Score',ascending=False).head(5)
hs
```

```python
hs.plot()
plt.ylabel('Amont of 10''s')
plt.xlabel('Countries')
plt.show()
```

<h4>This far does it seem that the guests from United Kingdom leave the highest score on their reviews and get the most responses from their review.</h4>

<h5>Q3, Based on the nights stayed at hotels, what is the average score per night from reviews by country</h5>

```python
avg = reviews[['Score','GuestCountry']].copy()
```

<h4>I make a copy of colum 'Score' and 'GuestCountry' and placed them in a new (dataframe) variable. This copy will make the base of the analasys I want to make. In the next cell, I will add another column that contains with the amount of nights stayed at hotels. This column is the one I made to_numeric before therefore this values already are int and able to be added mathematically</h4>

```python
avg['NumberOfNights'] = NumberOfNights
avg
```

<h4>After the adding of the column with nights stayed. I now want to reduce the rows and add all the values of the countries statistics into to a there unique country. When this is done, I can now once again add another column with my calculations of the averge score based of score per night stayed</h4>

```python
gva = avg.groupby(['GuestCountry']).sum()
gva['Average_score_per_night'] = gva['Score'] / gva['NumberOfNights']
```

```python
gva.sort_values('Average_score_per_night',ascending=False)
```

<h4>To make it more realistic, and more accurate scoring. The minimum of 14 nights will show a more just picture of how good the hotel have been.</h4>

```python
gva.loc[(gva['NumberOfNights'] >= 13)].sort_values('Average_score_per_night',ascending=False)
```

<h4>Based on the nights stayed at hotels sorted by county, and we divid the totalt review-score with the totalt amount of nights. The result differ alot from the previous results. The United Kingdom who had the most reviews with 10 out of 10, their average have now placed them at number 19 on the highest reviews, with an average value of 5.1</h4>
